```python
'''
zip_code = input("Enter zip-code\n")

# https://weather.com/weather/hourbyhour/l/02115:4:US

url_builder = zip_code + ":4:US"
url = "https://weather.com/weather/hourbyhour/l/"
url += url_builder

#url = "https://weather.com/weather/hourbyhour/l/02115:4:US"

page = requests.get(url)
'''

time = []
hourly_time = []
hourly_date = []
day = []
description = []
temp = []
feels = []
precip = []
humidity = []
wind = []
temp_dict = {}
feels_dict = {}
wind_dict = {}

url = "https://weather.com/weather/hourbyhour/l/02115:4:US"

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

closed_n = len(soup.find_all('tr', class_='closed'))

for i in range(closed_n):
    test = soup.find_all('tr', class_='closed')[i]
    hourly_time_ = test.find(class_="hourly-time").get_text()
    hourly_date_ = test.find(class_="hourly-date").get_text()
    description_ = test.find(class_="hidden-cell-sm description").get_text()
    temp_ = test.find(class_="temp").get_text()
    feels_ = test.find(class_="feels").get_text()
    precip_ = test.find(class_="precip").get_text()
    humidity_ = test.find(class_="humidity").get_text()
    wind_ = test.find(class_="wind").get_text()
    hourly_time.append(hourly_time_)
    hourly_date.append(hourly_date_)
    description.append(description_)
    temp.append(temp_)
    feels.append(feels_)
    precip.append(precip_)
    humidity.append(humidity_)
    wind.append(wind_)
    temp_dict[hourly_time_] = temp_
    feels_dict[hourly_time_] = feels_
    wind_dict[hourly_time_] = wind_




min_temp = min(temp_dict.items(), key=lambda x: x[1])
min_feels = min(feels_dict.items(), key=lambda x: x[1])
max_temp = max(temp_dict.items(), key=lambda x: x[1])
max_feels = max(feels_dict.items(), key=lambda x: x[1])
min_wind = min(wind_dict.items(), key=lambda x: x[1])
max_wind = max(wind_dict.items(), key=lambda x: x[1])

print("The minimum temperature in the next few hours is going to be: {0} at {1}.".format(min_temp[1], min_temp[0]))
print("The minimum 'feels like' in the next few hours is going to be: {0} at {1}.".format(min_feels[1], min_feels[0]))
print("The maximum temperature in the next few hours is going to be: {0} at {1}.".format(max_temp[1], max_temp[0]))
print("The maximum 'feels like' in the next few hours is going to be: {0} at {1}.".format(max_feels[1], max_feels[0]))
print("The minimum wind speed in the next few hours is going to be: {0} at {1}.".format(min_wind[1], min_wind[0]))
print("The maximum wind speed in the next few hours is going to be: {0} at {1}.".format(max_wind[1], max_wind[0]))

weather = pd.DataFrame({
        "hourly_time": hourly_time,
        "hourly_date": hourly_date,
        "description": description,
        "temp": temp, 
        "feels": feels, 
        "precip": precip,
        #"humidity": humidity,
        "wind":wind
    })

weather
```

```python
url2 = "https://weather.com/weather/tenday/l/02115:4:US"
page2 = requests.get(url2)
soup2 = BeautifulSoup(page2.content, 'html.parser')

time = []
hourly_time = []
hourly_date = []
day = []
description = []
temp = []
feels = []
precip = []
humidity = []
wind = []
high = []
low = []
date = []

closed_n2 = len(soup2.find_all('tr', class_='clickable closed'))


for i in range(closed_n2):
    test2 = soup2.find_all('tr', class_='clickable closed')[i]
    date_ = test2.find(class_="day-detail clearfix").get_text()
    day_ = test2.find(class_="date-time").get_text()
    description_ = test2.find(class_="description").get_text()
    temp_ = test2.find(class_="temp").get_text()
    splitter = temp_.find('°')
    high_ = temp_[0:splitter+1]
    low_ = temp_[splitter+1:]
    precip_ = test2.find(class_="precip").get_text()
    humidity_ = test2.find(class_="humidity").get_text()
    wind_ = test2.find(class_="wind").get_text()
    
    date.append(date_)
    day.append(day_)
    high.append(high_)
    low.append(low_)
    hourly_date.append(date_)
    description.append(description_)
    #temp.append(temp_)
    
    precip.append(precip_)
    humidity.append(humidity_)
    wind.append(wind_)
    
weather2 = pd.DataFrame({
        "date": date,
        "day": day,
        "description": description,
        #"temp": temp, 
        "high": high,
        "low": low,
        "precip": precip,
        "humidity": humidity,
        "wind":wind
    })
weather2
```
