```python
url2 = "https://weather.com/weather/tenday/l/02115:4:US"
page2 = requests.get(url2)
soup2 = BeautifulSoup(page2.content, 'html.parser')

time = []
hourly_time = []
hourly_date = []
day = []
description = []
temp = []
feels = []
precip = []
humidity = []
wind = []
high = []
low = []
date = []

closed_n2 = len(soup2.find_all('tr', class_='clickable closed'))


for i in range(closed_n2):
    test2 = soup2.find_all('tr', class_='clickable closed')[i]
    date_ = test2.find(class_="day-detail clearfix").get_text()
    day_ = test2.find(class_="date-time").get_text()
    description_ = test2.find(class_="description").get_text()
    temp_ = test2.find(class_="temp").get_text()
    splitter = temp_.find('°')
    high_ = temp_[0:splitter+1]
    low_ = temp_[splitter+1:]
    precip_ = test2.find(class_="precip").get_text()
    humidity_ = test2.find(class_="humidity").get_text()
    wind_ = test2.find(class_="wind").get_text()
    
    date.append(date_)
    day.append(day_)
    high.append(high_)
    low.append(low_)
    hourly_date.append(date_)
    description.append(description_)
    #temp.append(temp_)
    
    precip.append(precip_)
    humidity.append(humidity_)
    wind.append(wind_)
    
weather2 = pd.DataFrame({
        "date": date,
        "day": day,
        "description": description,
        #"temp": temp, 
        "high": high,
        "low": low,
        "precip": precip,
        "humidity": humidity,
        "wind":wind
    })
weather2
```
